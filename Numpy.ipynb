```python
import numpy as np

marks = np.random.randint(0, 101, 50)

print("Total =", np.sum(marks))
print("Average =", np.mean(marks))
print("Highest =", np.max(marks))
print("Lowest =", np.min(marks))
```

```python
import numpy as np

marks = np.random.randint(0, 101, 50)

print("Above 75 =", np.sum(marks > 75))
print("Below 35 =", np.sum(marks < 35))
print("Between 35 and 75 =",
      np.sum((marks >= 35) & (marks <= 75)))
```

```python
import numpy as np

marks = np.array([95, 80, 65, 50, 30])

for i in marks:
    if i >= 75:
        print(i, "A")
    elif i >= 60:
        print(i, "B")
    elif i >= 35:
        print(i, "C")
    else:
        print(i, "Fail")
```

```python
import numpy as np

marks = np.random.randint(35, 101, (5, 5))

print("Marks Matrix:")
print(marks)

print("Total of each student:")
print(np.sum(marks, axis=1))
```

```python
import numpy as np

marks = np.random.randint(35, 101, (5, 5))

print("Marks Matrix:")
print(marks)

print("Highest Marks in Each Subject:")
print(np.max(marks, axis=0))
```

```python
import numpy as np

marks = np.random.randint(35, 101, (5, 5))

for i in range(5):
    total = np.sum(marks[i])
    avg = np.mean(marks[i])
    per = avg

    if per >= 75:
        grade = "A"
    elif per >= 60:
        grade = "B"
    elif per >= 35:
        grade = "C"
    else:
        grade = "Fail"

    print("Student", i + 1)
    print("Marks =", marks[i])
    print("Total =", total)
    print("Average =", avg)
    print("Percentage =", per)
    print("Grade =", grade)
    print()
```
